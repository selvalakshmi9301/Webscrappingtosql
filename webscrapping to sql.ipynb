```python
from selenium import webdriver
from bs4 import BeautifulSoup 
```

```python
driver=webdriver.Firefox()
driver.get("https://www.flipkart.com/search?q=Mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
driver
```

```python
html_content=driver.page_source
html_content
```

```python
Soup=BeautifulSoup(html_content,'html.parser')
Soup
```

```python
Apple_flipcart=Soup.find_all('div',class_='_75nlfW')
print(len(Apple_flipcart))
```

```python
Name=[]
Ratings_Stars=[]
No_of_Ratings_and_Review=[]
Product_Description=[]
Price=[]
for i in Apple_flipcart:
       a=i.find('div',class_='KzDlHZ')
       Name.append(a.string)
       b=i.find('div',class_='XQDdHH')
       Ratings_Stars.append(b.string)
       c=i.find('span',class_='Wphh3N')
       No_of_Ratings_and_Review.append(c.string)
       d=i.find('div',class_='_6NESgJ')
       Product_Description.append(d.text)
       e=i.find('div',class_='Nx9bqj _4b5DiR')
       Price.append(d.string)
print(Name)    
print(Ratings_Stars)
print(No_of_Ratings_and_Review)
print(Product_Description)
print(Price)
       
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Mobile_Name':Name,'Ratings':Ratings_Stars,'RatingReviews':No_of_Ratings_and_Review,'Mobile_Description':Product_Description,'Cost_Price':Price})
df
```
